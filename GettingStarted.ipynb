Read the data back in and see what it looks like

```python
import pandas as pd
pd.set_option('display.max_columns', 200)

df = pd.read_csv('testing/2000_hands_000.csv')
df.head()
```

Column names fall into a few categories:
1. Each player's card power (e.g. p0c1 == Player 0's 1st card's power), and one-hot-encoded suit
2. Same info about the turn card (TC_power and one-hot-encoded suit)
3. Global information (trump suit, caller, round trump was called, if they went alone)
4. The winners of each trick
5. How many points each team got for that hand
6. The order of all cards played (name + suit + player-id, e.g. JH0 is the Jack of Hearts, played by player 0)
7. A string result of the hand (Single, Loner, Sweep, Euchre)
8. The "True ID" of each of the players -- can ignore these columns.

In the output datasets, Player 3 is always the dealer, making Player 0 the left-of-the-dealer, and so on.

----------

To do a bit of data science, let's look at answering a simple question -- how many points-per-call does each position get? E.g., when the dealer calls trump, what is the expected increase in their team's score?

```python
import numpy as np
for playerid in range(4):
    print('PLAYER %i' %playerid)
    sub = df[df['caller'] == playerid]
    col, opp_col = ['points02', 'points13'] if playerid%2 == 0 else ['points13', 'points02']
    ppc = (sum(sub[col]-sub[opp_col]))/len(sub)
    std = np.sqrt(np.std(sub[col])**2 + np.std(sub[opp_col])**2)

    print('Points-per-call: %.2f +/- %.2f' %(ppc, std))
    print()
```

Without much data (only 2000 hands), we can see that the uncertainties are high. If we were to run this again with more data, we would see some more precise estimates of the points-per-call for each player.
