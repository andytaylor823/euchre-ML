## Do the rest of the formatting on a HPC (see `format_data.sh`)

```python
expected_count_comma = 122
def count_odd_lines(infile):
    with open(infile, 'r') as f:
        lines = f.read()
    
    weird_lines = [[i, len(line.split(','))] for i, line in enumerate(lines.split('\n')) if len(line.split(',')) != expected_count_comma]
    
    return weird_lines
```

```python
for i in range(500, 797):
    file = 'RLdataset/10000_hands_' + str(i) + '.csv'
    weird_lines = count_odd_lines(file)
    if len(weird_lines) > 1:
        print(file+':', len(weird_lines), weird_lines[:5])
```

```python
for i in range(500, 797):
    file = 'RLdataset/10000_hands_' + str(i) + '.csv'
    weird_lines = count_odd_lines(file)
    if len(weird_lines) > 1:
        print('Re-doing file %s' %file)
        with open(file, 'r') as infile:
            dirty = infile.read()
        
        clean = '\n'.join([line for line in dirty.split('\n') if len(line.split(',')) == expected_count_comma])
        with open(file, 'w') as outfile:
            outfile.write(clean)
        
        weird_lines = count_odd_lines(file)
        if len(weird_lines) > 1:
            print('ERROR IN CLEANING FILE %s' %file)
```

```python
def get_mispredicts(model, test, prnt=True, every=int(1e4)):
    missed_data, missed_target, missed_pred = [], [], []
    skipcount = len(pd.read_csv('missed_target.csv')) if os.path.exists('missed_target.csv') else 0
    
    for i, (data, target) in enumerate(test):
        if i < skipcount:
            continue
        if i % every == 0:
            if i > 0:
                for filename, arr in zip(['missed_'+f+'.csv' for f in ['data', 'target', 'pred']], [missed_data, missed_target, missed_pred]):
                    with open(filename, 'a') as f:
                        for line in arr:
                            f.write(','.join([str(int(v)) for v in line]))
                            f.write('\n')
                return missed_data, missed_target, missed_pred
                missed_data, missed_target, missed_pred = [], [], []
            if prnt:
                if i != 0:
                    print('Predicting %i (dt = %.1f seconds)...' %(i, time.time()-t_checkpoint))
                else:
                    print('Predicting %i...' %i)
                t_checkpoint = time.time()
    
        
        ypred = model.predict(data)
        for i in range(len(ypred)):
            ypred[i] = [0 if ypred[i][j] != ypred[i].max() else 1 for j in range(len(ypred[i]))]
            correct = sum(ypred[i]-target[i]).numpy() == 0
            if not correct:
                missed_data.append(data[i].numpy())
                missed_target.append(target[i].numpy())
                missed_pred.append(ypred[i])
                
    print('Missed a total of %.1f%% (%i / %i)' %(len(missed_pred)/i*100, len(missed_pred), i))
    return missed_data, missed_target, missed_pred    
```

```python
def get_mispredicts_v2(model, test, prnt=True, every=int(1e4)):
    missed_data, missed_target, missed_pred = [], [], []
    skipcount = len(pd.read_csv('missed_target.csv'))/(0.02*10) if os.path.exists('missed_target.csv') else 0
    test = test.skip(int(skipcount))
    filenames = ['missed_'+f+'.csv' for f in ['data', 'target', 'pred']]
    t_checkpoint = time.time()
    
    for i, (data, target) in enumerate(test):
        if i == 0: print('Starting the loop...')
        #if i < skipcount: continue
        if i % every == 0:
            if len(missed_data) > 0:
                print('Writing out %i misses...' %len(missed_data))
                for filename, arr in zip(filenames, [missed_data, missed_target, missed_pred]):
                    with open(filename, 'a') as f:
                        for line in arr:
                            f.write(','.join([str(int(v)) for v in line]))
                            f.write('\n')
                missed_data, missed_target, missed_pred = [], [], []
            if prnt:
                if i != 0:
                    print('Predicting %i (dt = %.1f seconds)...' %(i, time.time()-t_checkpoint))
                else:
                    print('Predicting %i...' %i)
                t_checkpoint = time.time()
                    
        ypred = model.predict(data)
        corrects = ypred.argmax(axis=-1) == target.numpy().argmax(axis=-1)
        for j in range(len(corrects)):
            if not corrects[j]:
                missed_data.append(data[j])
                missed_target.append(target[j])
                missed_pred.append(ypred[j])
                #print('FOUND ONE: %i' %i)
    
    if len(missed_data) > 0:
        for filename, arr in zip(filenames, [missed_data, missed_target, missed_pred]):
            with open(filename, 'a') as f:
                for line in arr:
                    f.write(','.join([str(int(v)) for v in line]))
                    f.write('\n')
    return missed_data, missed_target, missed_pred
```
